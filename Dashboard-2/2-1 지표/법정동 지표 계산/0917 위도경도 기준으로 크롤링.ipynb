```python
url = "https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?x={}&y={}".format(lat, log)
headers = {"Authorization": "KakaoAK 1d53225a36103804a9bb5d0585243989"}
api_test = requests.get(url,headers=headers)
url_text = json.loads(api_test.text)
```

```python
region_type1_list = []
code_1_list = []
region_f_1_list = []
region_s_1_list = []
region_t_1_list = []

region_type2_list = []
code_2_list = []
region_f_2_list = []
region_s_2_list = []
region_t_2_list = []


for i in tqdm(range(len(bus))):
    lat = bus['xpos'][i]
    log = bus['ypos'][i]
        
    url = "https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?x={}&y={}".format(lat, log)
    headers = {"Authorization": "KakaoAK 1d53225a36103804a9bb5d0585243989"}
    api_test = requests.get(url,headers=headers)
    url_text = json.loads(api_test.text)
    
    try:
        region_type1 = url_text['documents'][0]['region_type']
        code_1 = url_text['documents'][0]['code']
        region_f_1 = url_text['documents'][0]['region_1depth_name']
        region_s_1 = url_text['documents'][0]['region_2depth_name']
        region_t_1 = url_text['documents'][0]['region_3depth_name']
    except:
        region_type1 = ''
        code_1 = ''
        region_f_1 = ''
        region_s_1 = ''
        region_t_1 = ''
        print('except! on 1')
    try:
        region_type2 = url_text['documents'][1]['region_type']
        code_2 = url_text['documents'][1]['code']
        region_f_2 = url_text['documents'][1]['region_1depth_name']
        region_s_2 = url_text['documents'][1]['region_2depth_name']
        region_t_2 = url_text['documents'][1]['region_3depth_name']
    except:
        region_type2 = ''
        code_2 = ''
        region_f_2 = ''
        region_s_2 = ''
        region_t_2 = ''
        print('except! on 2')
        
    region_type1_list.append(region_type1)
    code_1_list.append(code_1)
    region_f_1_list.append(region_f_1)
    region_s_1_list.append(region_s_1)
    region_t_1_list.append(region_t_1)

    region_type2_list.append(region_type2)
    code_2_list.append(code_2)
    region_f_2_list.append(region_f_2)
    region_s_2_list.append(region_s_2)
    region_t_2_list.append(region_t_2)
```

```python
B_H = pd.DataFrame(np.column_stack([region_type1_list, code_1_list, region_f_1_list, region_s_1_list, region_t_1_list,
                                               region_type2_list, code_2_list, region_f_2_list, region_s_2_list, region_t_2_list]), 
                               columns=['법정동', '법정동코드', 'B_시도명', 'B_시군구', 'B_동',
                                       '행정동', '행정동코드', 'H_시도명', 'H_시군구', 'H_동'])
B_H
```

```python
fin = pd.concat([bus, B_H], axis=1)
fin
```

```python
fin.to_csv('Fin_data/편의_편의점.csv', encoding='utf-8-sig')
```
