```python
import pandas as pd
import numpy as np
```

# 데이터 로드

```python
bus = pd.read_excel('1)재크롤링한 지표 데이터/최종3_좌표, 주소 데이터.xlsx', sheet_name='교통_버스정류소(1)')
bus = bus[bus['B_시도명'] =='서울특별시']
bus.sample()
```

```python
sub = pd.read_excel('1)재크롤링한 지표 데이터/최종3_좌표, 주소 데이터.xlsx', sheet_name='교통_지하철역(1)')
sub = sub[sub['B_시도명'] =='서울특별시']
sub.sample()
```

```python
pol = pd.read_excel('1)재크롤링한 지표 데이터/최종3_좌표, 주소 데이터.xlsx', sheet_name='안전_경찰서(1)')
pol.drop('청', axis=1, inplace=True)
pol.drop('서', axis=1, inplace=True)
pol = pol[pol['B_시도명'] =='서울특별시']
pol.sample()
```

```python
fire = pd.read_excel('1)재크롤링한 지표 데이터/최종3_좌표, 주소 데이터.xlsx', sheet_name='안전_소방서(1)')
fire = fire[fire['B_시도명'] =='서울특별시']
fire.sample()
```

```python
hos = pd.read_excel('1)재크롤링한 지표 데이터/최종3_좌표, 주소 데이터.xlsx', sheet_name='의료_병원(1)')
hos = hos[hos['B_시도명'] =='서울특별시']
hos.sample()
```

```python
medi = pd.read_excel('1)재크롤링한 지표 데이터/최종3_좌표, 주소 데이터.xlsx', sheet_name='의료_약국(1)')
medi = medi[medi['B_시도명'] =='서울특별시']
medi.sample()
```

```python
dept = pd.read_excel('1)재크롤링한 지표 데이터/최종3_좌표, 주소 데이터.xlsx', sheet_name='편의_백화점(1)')
dept.drop('대분류', axis=1, inplace=True)
dept = dept[dept['B_시도명'] =='서울특별시']
dept.sample()
```

```python
conv = pd.read_excel('1)재크롤링한 지표 데이터/최종3_좌표, 주소 데이터.xlsx', sheet_name='편의_편의점')
conv = conv[conv['B_시도명'] =='서울특별시']
conv.sample()
```

```python
park = pd.read_excel('1)재크롤링한 지표 데이터/최종3_좌표, 주소 데이터.xlsx', sheet_name='문화_공원(1)')
park = park[park['B_시도명'] =='서울특별시']
park.sample()
```

```python
the = pd.read_excel('1)재크롤링한 지표 데이터/최종3_좌표, 주소 데이터.xlsx', sheet_name='문화_영화관(1)')
the = the[the['B_시도명'] =='서울특별시']
the.sample()
```

# 법정동 세기

```python
print(bus.shape)
print(hos.shape)
print(the.shape)
print(conv.shape)
print(park.shape)
print(dept.shape)
print(medi.shape)
print(fire.shape)
print(pol.shape)
print(sub.shape)
```

```python
hos = hos[['B_시군구', 'B_동']]
hos['sub지표'] = '병원'

the = the[['B_시군구', 'B_동']]
the['sub지표'] = '영화관'

pol = pol[['B_시군구', 'B_동']]
pol['sub지표'] = '경찰서'

conv = conv[['B_시군구', 'B_동']]
conv['sub지표'] = '편의점'

medi = medi[['B_시군구', 'B_동']]
medi['sub지표'] = '약국'

fire = fire[['B_시군구', 'B_동']]
fire['sub지표'] = '소방서'

dept = dept[['B_시군구', 'B_동']]
dept['sub지표'] = '백화점'

park = park[['B_시군구', 'B_동']]
park['sub지표'] = '공원'

sub = sub[['B_시군구', 'B_동']]
sub['sub지표'] = '지하철'

bus = bus[['B_시군구', 'B_동']]
bus['sub지표'] = '버스'
```

```python
data = pd.concat([bus, hos, the, conv, park, dept, medi, fire, pol, sub], axis=0)#.shape
data
```

```python
a = data[data['B_시군구'] == '마포구']
a[a['B_동'] == '신정동']
```

```python
data['개수'] = 1

new_data = pd.pivot_table(data, index=['B_시군구', 'B_동'],
                          columns = ['sub지표'],
                          values=['개수'], aggfunc=np.sum, fill_value=0)

new_data.columns = ['_'.join(col) for col in new_data.columns.values]
new_data.columns
```

```python
new_data.columns = ['경찰서', '공원', '백화점', '버스', '병원', '소방서', '약국', '영화관', '지하철', '편의점']
new_data
```

```python
dist_df = new_data.reset_index(level=[0,1])
dist_df
```

```python
dist_df.to_csv('법정동별_지표_개수.csv', encoding='utf-8-sig')
```
