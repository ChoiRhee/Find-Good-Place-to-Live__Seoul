```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('법정동별_지표_개수.csv')
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
report = pd.read_excel('Report.xls')
report
```

```python
all_df = pd.read_excel('KIKmix.20210705.xlsx')
all_df = all_df[all_df['시도명'] == '서울특별시']
all_df
```

```python
def remove(text):
    if text == '제기동':
        pass
    elif '홍제' in text:
        text = '홍제'+text[3:]
    else:
        text = text.replace('제', '')
    return text
```

```python
# 해당 동에 인구밀도 넣기
all_df['읍면동명'].fillna('x', inplace=True)
all_df['제2읍면동명'] = all_df.apply(lambda x: remove(x['읍면동명']), axis=1)

merged_df = pd.merge(all_df, report, left_on = '제2읍면동명', right_on = '동')#, how='outer')
merged_df
```

```python
df = pd.DataFrame(merged_df['동'].value_counts()).reset_index()
df.columns = ['count_동', '개수']
df
```

```python
# 법정동 인구밀도 구하기
seoul_df = pd.merge(merged_df, df, left_on = '동', right_on = 'count_동', how='outer')
seoul_df['법정동 인구밀도'] = seoul_df['인구밀도'] / seoul_df['개수']

cal_df = pd.DataFrame(seoul_df.groupby('동리명')['법정동 인구밀도'].sum())
cal_df = cal_df.reset_index()
cal_df
```

```python
data_df = pd.merge(data, cal_df, left_on = 'B_동', right_on = '동리명', how='outer')
data_df.to_csv('Calculated_B_pop.csv', encoding='utf-8-sig')
```
