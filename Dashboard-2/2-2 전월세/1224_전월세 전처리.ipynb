```python
import pandas as pd
import numpy as np
df2016 = pd.read_csv('서울특별시_전월세가_정보_2016.csv',  encoding = 'cp949', sep='\;')
df2017 = pd.read_csv('서울특별시_전월세가_정보_2017.csv',  encoding = 'cp949', sep='\;')
df2018 = pd.read_csv('서울특별시_전월세가_정보_2018.csv',  encoding = 'cp949', sep='\;')
df2019 = pd.read_csv('서울특별시_전월세가_정보_2019.csv',  encoding = 'cp949', sep='\;')
df2020 = pd.read_csv('서울특별시_전월세가_정보_2020.csv',  encoding = 'cp949', sep='\;')

df2016
```

```python
# 2016 ~ 2020 합치기!
df = pd.concat([df2016, df2017, df2018, df2019, df2020])

# 칼럼명 변경
df.columns = ['지번코드', '기관코드', '일련번호', '접수년도', '자치구코드', '자치구명', '법정동코드', '법정동명', '본번', '부번', '건물명', '층', '임대건물코드', '임대건물명', '임대면적', '전월세구분코드', '전월세구분', '보증금', '임대료', '계약년도', '계약일자', '건축년도']

# 건축년도 연도만 남김
df['건축년도'] = [d[:4] for d in df['건축년도']]

# 칼럼명 변경
df.columns = ['지번코드', '기관코드', '일련번호', '접수년도', '자치구코드', '자치구명', '법정동코드', '법정동명', '본번', '부번', '건물명', '층', '임대건물코드', '임대건물명', '임대면적', '전월세구분코드', '전월세구분', '보증금', '임대료', '계약년도', '계약일자', '건축년도']

# 건축년도 연도만 남김
df['건축년도'] = [d[:4] for d in df['건축년도']]

# 사용 안할 칼럼 삭제
df.drop(['지번코드', '기관코드', '일련번호', '자치구코드', '법정동코드', '본번', '부번', '임대건물코드', '전월세구분코드', '접수년도', '건물명', '계약일자', '층'], axis = 1, inplace = True)

# index 재설정
df = df.reset_index(drop = True)

# 평당임대료 계산
df['평당임대료'] = df['임대료'] / df['임대면적']

df
```
