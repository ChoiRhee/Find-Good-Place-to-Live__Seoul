```python
import pandas as pd
import numpy as np
```

```python
# 데이터 로드하기
bu_2016 = pd.read_excel('2016_부동산실거래.xlsx', sheet_name='Sheet1')
bu_2016.head()
```

```python
bu_2017 = pd.read_excel('2017_부동산실거래.xlsx', sheet_name='Sheet1')
bu_2018 = pd.read_excel('2018_부동산실거래.xlsx', sheet_name='Sheet1')
bu_2019 = pd.read_excel('2019_부동산실거래.xlsx', sheet_name='Sheet1')
bu_2020 = pd.read_excel('2020_부동산실거래.xlsx', sheet_name='Sheet1')
```

```python
# 필요한 값들 정리
needed_list = ['자치구명', '법정동명', '신고년도', '대지권면적', '건물주용도', '물건금액', '건축년도']

bu_2016 = bu_2016[needed_list]
bu_2017 = bu_2017[needed_list]
bu_2018 = bu_2018[needed_list]
bu_2019 = bu_2019[needed_list]
bu_2020 = bu_2020[needed_list]
```

```python
# 0값들을 결측치로 대체
import warnings
warnings.filterwarnings(action='ignore') 

idx = bu_2016[bu_2016['대지권면적'] <=0].index
bu_2016.loc[idx, '대지권면적'] = np.nan
idx = bu_2016[bu_2016['건축년도'] <=0].index
bu_2016.loc[idx, '건축년도'] = np.nan

idx = bu_2017[bu_2017['대지권면적'] <=0].index
bu_2017.loc[idx, '대지권면적'] = np.nan
idx = bu_2017[bu_2017['건축년도'] <=0].index
bu_2017.loc[idx, '건축년도'] = np.nan

idx = bu_2018[bu_2018['대지권면적'] <=0].index
bu_2018.loc[idx, '대지권면적'] = np.nan
idx = bu_2018[bu_2018['건축년도'] <=0].index
bu_2018.loc[idx, '건축년도'] = np.nan

idx = bu_2019[bu_2019['대지권면적'] <=0].index
bu_2019.loc[idx, '대지권면적'] = np.nan
idx = bu_2019[bu_2019['건축년도'] <=0].index
bu_2019.loc[idx, '건축년도'] = np.nan

idx = bu_2020[bu_2020['대지권면적'] <=0].index
bu_2020.loc[idx, '대지권면적'] = np.nan
idx = bu_2020[bu_2020['건축년도'] <=0].index
bu_2020.loc[idx, '건축년도'] = np.nan
```

```python
a = pd.DataFrame(bu_2016.isnull().sum() / len(bu_2016) *100)
a.columns = ['2016']
b = pd.DataFrame(bu_2017.isnull().sum() / len(bu_2017) *100)
b.columns = ['2017']
c = pd.DataFrame(bu_2018.isnull().sum() / len(bu_2018) *100)
c.columns = ['2018']
d = pd.DataFrame(bu_2019.isnull().sum() / len(bu_2019) *100)
d.columns = ['2019']
e = pd.DataFrame(bu_2020.isnull().sum() / len(bu_2020) *100)
e.columns = ['2020']
```

```python
# 결측치 제거
bu_2016 = bu_2016.drop('대지권면적', axis=1)#[bu_2017['건축년도'] != np.nan]
bu_2016 = bu_2016.dropna(axis=0)

bu_2017 = bu_2017.drop('대지권면적', axis=1)#[bu_2017['건축년도'] != np.nan]
bu_2017 = bu_2017.dropna(axis=0)

bu_2018 = bu_2018.drop('대지권면적', axis=1)#[bu_2017['건축년도'] != np.nan]
bu_2018 = bu_2018.dropna(axis=0)

bu_2019 = bu_2019.drop('대지권면적', axis=1)#[bu_2017['건축년도'] != np.nan]
bu_2019 = bu_2019.dropna(axis=0)

bu_2020 = bu_2020.drop('대지권면적', axis=1)#[bu_2017['건축년도'] != np.nan]
bu_2020 = bu_2020.dropna(axis=0)
```

```python
# 건축년도 변환
def trans_year(year):
    if year >= 2010:
        return '2010s'
    elif year >= 2000:
        return '2000s'
    elif year >= 1990:
        return '1990s'
    else:
        return '기타'
```

```python
bu_2016['건축년도'] = bu_2016.apply(lambda x: trans_year(x['건축년도']), axis=1)
bu_2017['건축년도'] = bu_2017.apply(lambda x: trans_year(x['건축년도']), axis=1)
bu_2018['건축년도'] = bu_2018.apply(lambda x: trans_year(x['건축년도']), axis=1)
bu_2019['건축년도'] = bu_2019.apply(lambda x: trans_year(x['건축년도']), axis=1)
bu_2020['건축년도'] = bu_2020.apply(lambda x: trans_year(x['건축년도']), axis=1)
```

```python
bu_2016 = pd.DataFrame(bu_2016.groupby(['자치구명', '법정동명','건물주용도', '건축년도'])['물건금액'].mean())
bu_2016 = bu_2016.reset_index()
bu_2016['신고년도'] = 2016

bu_2017 = pd.DataFrame(bu_2017.groupby(['자치구명', '법정동명','건물주용도', '건축년도'])['물건금액'].mean())
bu_2017 = bu_2017.reset_index()
bu_2017['신고년도'] = 2017

bu_2018 = pd.DataFrame(bu_2018.groupby(['자치구명', '법정동명','건물주용도', '건축년도'])['물건금액'].mean())
bu_2018 = bu_2018.reset_index()
bu_2018['신고년도'] = 2018

bu_2019 = pd.DataFrame(bu_2019.groupby(['자치구명', '법정동명','건물주용도', '건축년도'])['물건금액'].mean())
bu_2019 = bu_2016.reset_index()
bu_2019['신고년도'] = 2019

bu_2020 = pd.DataFrame(bu_2020.groupby(['자치구명', '법정동명','건물주용도', '건축년도'])['물건금액'].mean())
bu_2020 = bu_2020.reset_index()
bu_2020['신고년도'] = 2020
```

```python
all_df = pd.concat([bu_2016, bu_2017, bu_2018, bu_2019, bu_2020], axis=0)
all_df
```

```python
all_df.reset_index(drop=True, inplace=True)
all_df.drop('index', axis=1, inplace=True)
```
